Sentence cleaning is a fundamental and crucial step in any natural language processing (NLP) problem. It involves several important tasks, including word tokenization, removal of special characters, stop words elimination, and lemmatization.

During word tokenization, the sentence is split into individual words, allowing us to analyze and process each word separately. Special characters, such as punctuation marks and symbols, are removed to focus on the meaningful content of the text.

Stop words, which are commonly used words with little semantic value (e.g., "the," "is," "and"), are eliminated to reduce noise and improve the quality of the text representation. Removing stop words helps to extract more relevant and meaningful information from the text.

Additionally, lemmatization is applied to convert words into their base or dictionary form (lemma). This process reduces word variations and ensures consistent representation. For example, words like "running," "ran," and "runs" would be transformed to their base form "run."

By performing these cleaning steps, we can preprocess the text data and create a cleaner and more consistent representation, which is essential for various NLP tasks such as text classification, information retrieval, sentiment analysis, and machine translation.

Overall, sentence cleaning plays a vital role in enhancing the quality and effectiveness of NLP models by reducing noise, improving text representation, and facilitating better understanding and analysis of the text data.

```python
# To reduce a word to its base or dictionary form
lematizer = WordNetLemmatizer()

def get_word_net_loc(t):
    if t.startswith('J'):
        return wordnet.ADJ
    elif t.startswith('V'):
        return wordnet.VERB
    elif t.startswith('N'):
        return wordnet.NOUN
    elif t.startswith('R'):
        return wordnet.ADV
    else:
        return None

# To perform lemmatization:
    
def lematize_txt(txt):
    lematizer = WordNetLemmatizer()
    lem_txt = []
    for word in txt:
        lem_txt.append(lematizer.lemmatize(word))
    return lem_txt

# To make a clean version of the text through several preprocessing steps:

def clean_corpus(input_df):
    df_length = input_df.shape[0]
    for idx, row in input_df.iterrows():
        # To remove punctuations and only have word or spaces
        remove_special_character = re.sub(r'[^\w\s]','', row["corpus"])

        # To tokenize the corpus
        corpus_token = nltk.word_tokenize(remove_special_character)

        # To remove all stop words
        stop_wrds = set(stopwords.words("english"))
        corpus_without_stopwrds = [word for word in corpus_token if word not in stop_wrds]

        # To lemmatize the words
        cleaned_corpus = lematize_txt(corpus_without_stopwrds)

        # cleaned_corpus = lematize_txt(nltk.pos_tag(corpus_without_stopwrds))
        
        # To join the words back
        cleaned_corpus = ' '.join(cleaned_corpus)
        df.loc[idx, "corpus"] = cleaned_corpus
        
        if not idx%2000:
            print(f"{idx}/{df_length}")
    
    return df
    
```

```python
# To clean the corpus
df = clean_corpus(df)
```

## Model Building

```python
# To import scikit-learn library modules

from sklearn.feature_extraction.text import TfidfTransformer, CountVectorizer, TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import recall_score, precision_score, f1_score, confusion_matrix, accuracy_score
```

```python
# Extracting the 'corpus' column from the DataFrame and assigning it to X_train.
X_train = df['corpus']

# Extracting the 'label' column from the DataFrame and assigning it to Y_train.
Y_train = df['label']
```

#### Machine Learning Models

```python
# To collect all unique words vector

countvector = CountVectorizer(max_features=500)
countvector.fit(X_train)

# To calculate frequency for each word in the document

frq_trm_mtrx = countvector.transform(X_train)

# To calculate inverse term frequency

tfidf = TfidfTransformer(norm="l2")
tfidf.fit(frq_trm_mtrx)
tfidf_mtrx = tfidf.transform(frq_trm_mtrx)
```

```python
# To define dictionary with machine learning models

model = {
    "RandomForest": RandomForestRegressor(n_estimators=100,max_depth=10, random_state=123),
    "LR" : LogisticRegression(),
    "naive_bayes" : GaussianNB(),
    "KNN" : KNeighborsClassifier()}
```

```python
# To perform train-test splitting of the data

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(tfidf_mtrx, Y_train, random_state=0)
```

```python
# To calculate the number of rows and columns to display ML models in a grid-like arrangement

no_model = len(model)
no_row = (no_model + 1) // 2  # +1 to accommodate uneven number of models
no_col = 2

# To set the figure size of object

fig = plt.figure(figsize=(12, 8))

# To create model directory if it does not exist

if not os.path.exists("models"):
    os.mkdir("models")

# To train a collection of models and calculate the confustion matrix (cm)

for idx, m in enumerate(model):
    model[m].fit(X_train.toarray(), y_train)
    # save model
    dump(model[m], f"models/{m}.pkl")
    
    y_prediction = model[m].predict(X_test.toarray())
    y_prediction = np.where(y_prediction>0.5, 1, 0)

    print("Metric - {}".format(m))
    print("Recall         : ", recall_score(y_pred=y_prediction, y_true=y_test))
    print("Precision      : ", precision_score(y_pred=y_prediction, y_true=y_test))
    print("F1 score       : ", f1_score(y_pred=y_prediction, y_true=y_test))
    print("Accuracy score : ", accuracy_score(y_pred=y_prediction, y_true=y_test))
    print("")
    
    cm = confusion_matrix(y_pred=y_prediction, y_true=y_test)
    
    # To plot the confusion matrix (cm)
    ax = fig.add_subplot(no_row, no_col, idx + 1)
    sns.heatmap(cm, cmap='Blues',ax=ax, annot=True, annot_kws={"fontsize": 10, "ha": "center"}, fmt=".2f")
    ax.set_title("Confusion Matrix - {}".format(m))
    ax.set_xlabel("Predicted")
    ax.set_ylabel("Actual")
    ax.set_xticks([0.5, 1.5])
    ax.set_yticks([0.5, 1.5])
    ax.tick_params(pad=5)
    ax.set_xticklabels(["Class 0", "Class 1"])
    ax.set_yticklabels(["Class 0", "Class 1"])
    ax.grid(False)

plt.tight_layout()
plt.show()
```

```python
# Extracting the "corpus" and "label" columns from the DataFrame df
X, y = df["corpus"], df["label"]

# Splitting the data into training and testing sets using a random seed of 0
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

# Converting the training data (corpus) into a list
corp = X_train.tolist()

# Converting the testing data (corpus) into a list
corp_test = X_test.tolist()

# Defining the vocabulary size for text processing as 2000
vocabsize = 2000
```

```python
# Creating one-hot encodings for the 'corp' dataset using a vocabulary size 'vocabsize'.
onehot_corp1 = [one_hot(words,vocabsize)for words in corp]

# Creating one-hot encodings for the 'corp_test' dataset using the same vocabulary size 'vocabsize'.
onehot_corp1_test = [one_hot(words,vocabsize)for words in corp_test]

# Padding Sentences to make them of same size
embedded_doc = pad_sequences(onehot_corp1, padding='post', maxlen=50)
embedded_doc_test = pad_sequences(onehot_corp1_test, padding='post', maxlen=50)

# Convert docs to numpy array
X_train = np.array(embedded_doc)
y_train = np.array(y_train)
X_test = np.array(embedded_doc_test)
y_test = np.array(y_test)
```

```python
# Define model architecture
# To use embedding layers with LSTM
model = Sequential()
model.add(Embedding(vocabsize,100,input_length=50,name="input"))
model.add(Dropout(0.2))
model.add(LSTM(100))
model.add(Dropout(0.2))
model.add(Dense(32,activation='tanh'))
model.add(Dropout(0.2))
model.add(Dense(1,activation='sigmoid', name="target"), )
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['binary_crossentropy'])
print(model.summary())
```

```python
# Define a generator function called bootstrap_generator that generates batches of data.
# This generator is designed for training purposes and helps with resampling data.

def bootstrap_generator(batch_size):
    # Create an infinite loop to keep generating batches continuously.
    while True:
        # Randomly select batch_size indices from the training data.
        batch_idx = np.random.choice(np.array(X_train).shape[0], batch_size)
        
        # Yield a tuple containing input and target data for the current batch.
        # The input data is a dictionary with a key "input_input," containing input samples.
        # The target data is a dictionary with a key "target," containing corresponding target labels.
        yield ({"input_input": np.array(X_train)[batch_idx]},
               {"target": np.array(y_train)[batch_idx]})
```

```python
# Define the name of the model, which will be used for saving the trained model.
model_name="LSTM"

# Create a ModelCheckpoint callback to save the best model weights during training.
# - The model will be saved to the 'models/' directory with the name '{model_name}.h5'.
# - Monitoring 'loss' to save the best model based on minimizing loss.
# - 'save_best_only=True' ensures only the best model is saved.
# - 'verbose=1' displays progress information.
modelCheckpoint = ModelCheckpoint(filepath=f"models/{model_name}.h5", monitor='loss', mode='min', save_best_only=True, verbose=1)

# Create an EarlyStopping callback to stop training if loss stops improving.
# - Monitoring 'loss' for improvement.
# - 'patience=10' allows training to continue for up to 10 epochs without improvement.
# - 'verbose=1' displays messages when early stopping is triggered.
# - 'restore_best_weights=True' restores the model to the best weights found during training.
earlystopping = EarlyStopping(monitor="loss", patience=10, verbose=1, restore_best_weights=True)

# Define the batch size for training data.
batch_size=512

# Train the model using a custom data generator 'bootstrap_generator'.
# - Training for a maximum of 100 epochs.
# - 'steps_per_epoch=10' specifies the number of batches to complete an epoch.
# - Callbacks include ModelCheckpoint and EarlyStopping to save the best model and stop early if needed.
model.fit(bootstrap_generator(batch_size), epochs=100, steps_per_epoch=10, callbacks=[modelCheckpoint, earlystopping])

```

```python
# Predict output
y_prediction = model.predict(X_test)
```

```python
# Create a subplot for a single plot in a figure with a specified size of 4x4 inches
f, ax = plt.subplots(1,1, figsize=(4,4))
y_prediction = np.where(y_prediction>0.5, 1, 0)
print("Metric - {}".format("Sequential NN"))
print("Recall         : ", recall_score(y_pred=y_prediction, y_true=y_test))
print("Precision      : ", precision_score(y_pred=y_prediction, y_true=y_test))
print("F1 score       : ", f1_score(y_pred=y_prediction, y_true=y_test))
print("Accuracy score : ", accuracy_score(y_pred=y_prediction, y_true=y_test))

print("")

cm = confusion_matrix(y_pred=y_prediction, y_true=y_test)

# Plot the confusion matrix
sns.heatmap(cm, cmap='Blues', annot=True, annot_kws={"fontsize": 10, "ha": "center"}, fmt=".2f")
ax.set_title("Confusion Matrix - {}".format("Sequential NN"), fontsize=8)
ax.set_xlabel("Predicted")
ax.set_ylabel("Actual")
ax.set_xticks([0.5, 1.5])
ax.set_yticks([0.5, 1.5])
ax.tick_params(pad=5)
ax.set_xticklabels(["Class 0", "Class 1"])
ax.set_yticklabels(["Class 0", "Class 1"])
ax.grid(False)
plt.tight_layout()
plt.show()
```
